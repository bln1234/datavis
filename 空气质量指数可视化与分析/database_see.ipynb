```python
import sqlite3
database="AQI_data.db"
conn = sqlite3.connect(database)
cursor = conn.cursor()
cursor.execute("SELECT * FROM AQI_data")
rows = cursor.fetchall()
for row in rows:
    print(row)
conn.close()

```

```python
import sqlite3

database = "AQI_data.db"
conn = sqlite3.connect(database)
cursor = conn.cursor()

# 假设要查看的表名是 'aqi_data'
table_name = 'AQI_data'

# 获取表的结构信息
cursor.execute(f"PRAGMA table_info({table_name})")
columns = cursor.fetchall()

# 提取字段名
column_names = [column[1] for column in columns]
print("Column names in the table:", column_names)

conn.close()
```
